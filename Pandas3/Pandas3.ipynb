## mpg.csv dosyasını okuyun ve df değişkenine atayın.

```python
import pandas as pd
df=pd.read_csv('mpg.csv')
df
```

## df'in içindeki sütunların birbirleriyle olan korelasyonlarına(ilişkilerine) bakın.

```python
df.corr()
```

## mpg sütunu ile diğer sütunların arasındaki korelasyonları özel olarak inceleyin.

```python
df.corr()['mpg']
```

## cylinders sütununda hangi unique değerler olduğuna liste halinde bakın.

```python
df['cylinders'].unique()
```

## cylinders sütunundaki unique değerlerin her birinden kaçar adet olduğuna bakın.

```python
df['cylinders'].value_counts()
```

## Arabanın modeli 75-80 yılları arasındaki arabaların hepsini getirin.(75 ve 80 dahil)

```python
df[df['model_year'].between(75,80)]
```

## horsepower'ı 155-180 arasındaki arabaların hepsini getirin.(155 dahil, 180 dahil değil) 

```python
df = df.drop(df[df['horsepower'] == '?'].index)
df['horsepower']=df['horsepower'].astype(float)
df[df['horsepower'].between(155,180,inclusive = 'left')]
```

## Elinizdeki df'i horsepower'a göre büyükten küçüğe sıralayın ardından df'in indexlerini sıfırlayın.

```python
df.sort_values(by=['horsepower'],ascending=False).reset_index(drop=True)
```

## Elinizdeki df'den en küçük 5 acceralation'a sahip olan arabaları getirin.

```python
df.sort_values(by=['acceleration'],ascending=True).head(5)
```

## CIA_Country_Facts.csv dosyasını okuyun ve yeniden df değişkenine atayın.(Artık eski df ile işimiz kalmadığı için aynı ismi verebiliriz.)

```python
##her ihtimale karşı ben farklı bir df açtım
df2=pd.read_csv('CIA_Country_Facts.csv')
df2
```

## Elimizdeki CIA'in ülkeler hakkında topladığı basit açık kaynak verileri df'inde kaç ülkenin verisi olduğunu bulun.

```python
df2.shape[0]
```

## Elimizdeki sütunlardan hangilerinde kaçar tane nan veri olduğunu bulun.

```python
df2.isnull().sum()
```

## Elimizde tamamen nan değerlerden oluşan bir satır varsa o satırı düşürün.(Tamamen nan değerlerden oluşan bir satır olup olmadığını, herhangi bir satırın düşüp düşmediğini kontrol edin)

```python
df2 = df2.dropna(how='all')
df2.shape[0]
```

## Her Bölgedeki(Region) toplam kişi sayısına bakın.

```python
df2.groupby('Region')['Population'].sum()
```

## Climate değerleri nan olanlar, tüm df'in ortalama climate değeriyle doldurun.(Gerçek bir analiz yapıyor olsaydık bu çok mantıklı bir hareket olmazdı, belki her bulunduğu bölgenin ortalama climate değeriyle doldurmak daha mantıklı bir hareket olurdu. Fakat şimdilik bu kadar kasmaya gerek yok. nan değerleri tüm climate ortalamalarıyla doldursanız yeterli.)

```python
df2['Climate'] = df2.groupby('Region')['Climate'].apply(lambda x: x.fillna(x.mean()))
df2['Climate'].isna().sum()
```

## Bölgeler ve Ülkelere göre gruplama yapın ardından Doğum miktarlarına bakın.

```python
df2.groupby(['Region','Country'])['Birthrate'].sum()
```

## Veriyi istediğiniz şekilde inceleyerek bu zamana kadar öğrendiğiniz bilgilerle anlamlı 5 adet yargı cümlesi çıkartın. ( En çok insan Asya kıtasında yaşıyor, En çok insan Afrinın sahralarında ölüyor, Ortalama en yüksek telefon sahibi Kuzey Amerikada, En çok okuma yazma oranına sahip ülkeler Baltık ülkeleri vb.)

```python
# En çok göç alan bölge
df2.groupby('Region')['Net migration'].sum().sort_values(ascending=False).head(1)
```

```python
# En az kişi başı gelire sahip bölge
df2.groupby('Region')['GDP ($ per capita)'].sum().sort_values(ascending=True).head(1)
```

```python
# Yüzdelik olarak en çok ekili tarım arazisine sahip bölge
df2.groupby('Region')['Crops (%)'].sum().sort_values(ascending=False).head(1)
```

```python
# En yoğun nüfusun bulundnuğu bölge
df2.groupby('Region')['Pop. Density (per sq. mi.)'].sum().sort_values(ascending=False).head(1)
```

```python
#En büyük kıyı uzunluğuna sahip bölge
df2['CoastLength']=df2['Coastline (coast/area ratio)']*df2['Area (sq. mi.)']
df2.groupby('Region')['CoastLength'].sum().sort_values(ascending=False).head(1)
```
